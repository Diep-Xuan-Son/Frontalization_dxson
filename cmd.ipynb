```python
import os
import cv2
import shutil

# img = cv2.imread('/home/son/AI/frontalization/data1/Facebook/ids3133/ids3133_1.jpg')
# cv2.imshow('a',img)
# cv2.waitKey()
for root, dirs, files in os.walk('/home/son/AI/frontalization/data1/Facebook/ids3133'):
	for file in files:
		if file.endswith(('.jpg', '.JPG')):
			print(root)
			if len(root.split('/')) < len('/home/son/AI/frontalization/data/Facebook/ids'.split('/')):
				with open(os.path.join(root, file), 'rb') as f:
					check_chars = f.read()[-2:]
				if check_chars != b'\xff\xd9':
					print('Not complete image-----------------------------------------------------')
					os.remove(os.path.join(root, file))
					file_new = file.split('.')[0] + '/' + file.split('.')[0] + '_1.jpg'
					shutil.copy2(os.path.join(root, file_new), os.path.join(root, file))
				continue
			path = os.path.join(root, file)
			with open(path, 'rb') as f:
				check_chars = f.read()[-2:]
			if check_chars != b'\xff\xd9':
				print('Not complete image-----------------------------------------------------')
				os.remove(path)
				continue
			print(path)
			img = cv2.imread(path)
			dim = (128, 128)
			# resize image
			resided = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
			cv2.imwrite(path,resided)
```
